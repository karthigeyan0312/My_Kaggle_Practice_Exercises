```python
import nltk
from nltk import word_tokenize

import wikipedia

input_text = wikipedia.summary("Lata Mangeshkar")
print(input_text)
```

```python
sentences = nltk.tokenize.sent_tokenize(input_text)
sentence_count = len(sentences)
print("Number of Sentences: ", sentence_count)
```

```python
print(sentences)
```

```python
words = nltk.tokenize.word_tokenize(input_text)
word_count = len(words)
print("Number of Words: ", word_count)
```

```python
print(words)
```

```python
from string import punctuation
punct_free_words = [word for word in words if word not in punctuation and not word.isdigit()]
clean_word_count = len(punct_free_words)
print("Number of Words(cleaned): ", clean_word_count)
```

```python
print(punct_free_words)
```

```python
print(nltk.pos_tag(punct_free_words))
```

```python
import spacy

pos_tags = spacy.load('en_core_web_sm')
doc = pos_tags(input_text)

print([(token.text, token.pos_) for token in doc])
```

```python
print([(token.text, token.label_) for token in doc.ents])
```

```python
import syllapy

syl_cnt_list = []

for word in words:
    count = syllapy.count(word)
    syl_cnt_list.append(count)

print(len(syl_cnt_list))
```

```python
syllable_count = 0

for i in range(0, len(syl_cnt_list)):
    syllable_count = syllable_count + syl_cnt_list[i]

print("Number of Syllables: ", syllable_count)
```

## Flesch Reading Ease Test:

> Higher scores indicate material that is easier to read, lower numbers mark passages that are more difficult to read.

$
    \begin{equation}
        FRES = 206.835 - 1.015 * \Bigg(\frac{total words}{total sentences}\Bigg) - 84.6 * \Bigg(\frac{total syllables}{total words} \Bigg)
    \end{equation}
$

|Score|School level (US)|Notes|
|---|---|---|
|100.00–90.00|5th grade|Very easy to read. Easily understood by an average 11-year-old student.|
|90.0–80.0|6th grade|Easy to read. Conversational English for consumers.|
|80.0–70.0|7th grade|Fairly easy to read.|
|70.0–60.0|8th & 9th grade|Plain English. Easily understood by 13- to 15-year-old students.|
|60.0–50.0|10th to 12th grade|Fairly difficult to read.|
|50.0–30.0|College|Difficult to read.|
|30.0–10.0|College graduate|Very difficult to read. Best understood by university graduates.|
|10.0–0.0|Professional|Extremely difficult to read. Best understood by university graduates.|

```python
fres = (206.835 - 1.015 * (word_count/sentence_count)) - (84.6 * (syllable_count/word_count))
print("Flesch Reading Ease Score: ", fres)
```

## Flesch-Kincaid Grade Level

> Presents a score as a U.S. grade level. 
The score does not have an upper bound.

$
    \begin{equation}
        Grade Level = 0.39 * \Bigg(\frac{total words}{total sentences}\Bigg) + 11.8 * \Bigg(\frac{total syllables}{total words} \Bigg) - 15.59
    \end{equation}
$

```python
gl = 0.39 * (word_count/sentence_count) + 11.8 * (syllable_count/word_count)
print("Flesch-Kincaid Grade Level: ", gl)
```
